```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
kl = cv2.imread("123.jpg")
kl = cv2.cvtColor(kl,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(kl)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
kls = cv2.cvtColor(kl, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(kls,cmap="gray")
plt.axis("off")
plt.title("Image to GrayScale Image ")
plt.show()
```

```python
imk = cv2.bitwise_not(kls)
plt.figure(figsize=(8,8))
plt.imshow(imk,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
ims = cv2.GaussianBlur(imk, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(ims,cmap="gray")
plt.axis("off")
plt.title("Smoothened Image")
plt.show()
```

```python
vr = cv2.divide(kls, 255 - ims, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(vr,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
